### One Hot Encoding - variables with many categories
one hot encoding is used for converting categorical data into numerical data

```python
import pandas as pd
import numpy as np

# loading mercedes dataset
data = pd.read_csv("mercedes.csv", usecols = ['X1', 'X2', 'X3', 'X4', 'X5', 'X6'])
data.head()
```

```python
# lets check how many unique values values are present in each column

for col in data.columns:
    print(col, ':', len(data[col].unique()), 'labels')
```

```python
# let's examine how many columns we will obtain after one hot encoding these variables

pd.get_dummies(data, drop_first = True).shape
```

We can see that from just 6 initial categorical variables, we end up with 117 new variables.

```python
# let's find the top 10 most frequent categories for the variables x2

data.X2.value_counts().sort_values(ascending = False).head(20)
```

```python
# lets make a list with the most frequent categoris for the variabls

top_10 = [x for x in data.X2.value_counts().sort_values(ascending = False).head(10).index]
top_10
```

```python
# and now we make the 10 binary variables

for label in top_10:
    data[label] = np.where(data['X2'] == label, 1, 0)
data[['X2'] + top_10].head(20)
```

```python
import pandas as pd
import numpy as np

# Read the file again
data = pd.read_csv('mercedes.csv', usecols=['X1', 'X2', 'X3', 'X4', 'X5', 'X6'])

# Determine the top 10 most frequent labels in the 'X2' column
top_10 = data['X2'].value_counts().head(10).index

# Function to create the dummy variables for the most frequent labels
def one_hot_top_x(df, variable, top_x_labels):
    for label in top_x_labels:
        df[variable + '_' + label] = np.where(df[variable] == label, 1, 0)

# Encode 'X2' into the 10 most frequent categories
one_hot_top_x(data, 'X2', top_10)
data.head()

    
 
```

```python
# find the 10 most frequent categoris for x1
top_10 = [x for x in data.X1.value_counts().sort_values(ascending = False).head(10).index]

# now create the 10 most frequent dummy variables for x1
one_hot_top_x(data, 'X1', top_10)
data.head()
```
